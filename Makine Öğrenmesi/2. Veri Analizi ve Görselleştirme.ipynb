### Veriyi kullanırken filtrelemek gerekebiliyor. Aşağıdaki kodda ortalama çanak uzunluğu ve genişliğinden büyük olan verilerin listelenmesi gösterilmektedir.

```python
filtre1 = veri.SepalLengthCm > 5.843333 # Çanak uzunluğu ortalamanın üstünde olan verileri filtrele
filtre2 = veri.SepalWidthCm > 3.057333 # Çanak genişliği ortalamanın üstünde olan verileri filtrele

yeniVeri = veri[filtre1 & filtre2]
yeniVeri.head(30)
```

### Veri setinde herhangi bir eksik datanın olup olmadığını kontrol etmek için aşağıdaki kod yazılabilir.

```python
veri.isnull().sum()
```

### Veri setindeki herhangi gereksiz bir sütunu kaldırmak için aşağıdaki kod yazılabilir.

```python
yeniVeri1 = veri.drop(['Id'],axis=1) # axis=1 sütunu(dikey eksen), axis=0 satırı(yatay eksen) belirtir.
yeniVeri1.head()
```

### Herhangi bir sütundaki tüm verileri çekmek için aşağıdaki kodu kullanabiliriz. Bu kod species sınıfını çekmek için kullanılır.

```python
yeniVeri1.Species
```

### Herhangi bir sütundaki tüm **eşsiz** verileri çekmek için aşağıdaki kodu kullanabiliriz.

```python
yeniVeri1.Species.unique()
```

### Herhangi bir sütunda eşsiz olan verilerin kaçar tane olduğunu tespit etmek için aşağıdaki kodları kullanabiliriz. Bu veri analizinde çok kullanılan bir yöntemdir.

```python
yeniVeri1.Species.value_counts()
```

### Üstteki işlemi aynı şekilde **groupby** metoduyla da yapabiliriz.

```python
yeniVeri1.groupby("Species").size()
```

### Buradan sonra veriyi görselleştirmeye başlayacağız. Verileri görselleştirirken **Seaborn** ve **Pyplot** kütüphanelerini sıklıkla kullanacağız.

### Aşağıda Seaborn kütüphanesinden **countplot** fonksiyonu gösterilmiştir. Bu fonksiyon bir çok parametre alabiliyor. Bu parametreleri incelemekte fayda var.

```python
sns.countplot(x="Species",data=veri)
plt.show() # Grafiği ekrana getirme fonksiyonu
```

### Eğer bu veriyi pasta grafiği halinde çizdirmek istersek aşağıdaki komutu da kullanabiliriz.

```python
veri["Species"].value_counts().plot.pie(autopct = '%1.1f%%', figsize=(10,8)) # autopct yüzdelik dilimleri yazdırmak için, figsize görselin en boy oranını ayarlar.
plt.ylabel('') # Gereksiz etiketi kaldırma
plt.title('') # Gereksiz başlığı kaldırma
plt.show()
```

### Yine aynı şekilde seaborn kütüphanesinin **catplot** fonksiyonunu kullanarak da grafikler çizilebilir. Bu fonksiyonun da bir çok parametresi bulunmakta.

```python
sns.catplot(x="Species", hue="Species", kind='count', data=veri) # hue verilen sınıfın eşsiz elemanlarını farklı renkte göstermeyi sağlar. kind çizim türünü belirtir.
plt.show()
```

### **catplot** fonksiyonu ile çok çeşitli grafikler de çizilebilir. Aşağıdaki kodda Petal uzunluğunun türlere göre grafiği çizdirilmiştir.

```python
sns.catplot(x="PetalLengthCm", hue="Species", kind='count', data=veri, aspect=2.5) # aspect boyutlandırma için kullanılmıştır.
plt.show()
```

### **Matplotlib** kütüphanesinin **bar** fonksiyonu aracılığıyla bir çubuk grafiği de oluşturabiliriz. Aşağıdaki kodda iris çiçeğinin farklı türlerine göre taç yaprağının uzunluklarını gösteren bir grafik çizilmiştir.

```python
plt.bar(veri["Species"], veri["PetalLengthCm"])
plt.show()
```

### **plot** fonksiyonu kullanılarak da veriler karşılaştırılabilir. Aşağıdaki kodda iris çiçeklerinin tüm uzunlukları grafikle çizilmiştir.

```python
yeniVeri1.plot()
plt.show()
```

### Veri setindeki verilerin daha kolay kullanımı için o veriyi bir değişkene atayabiliriz.

```python
versicolor = veri[veri.Species=="Iris-versicolor"]
virginica = veri[veri.Species=="Iris-virginica"]
setosa = veri[veri.Species=="Iris-setosa"]
```

### Bu değişkenleri **plot** fonksiyonuyla kullanalım. Aşağıdaki kodda iris çiçek türlerinin ID'leri ile taçyapraklarının uzunluklarını görselleştiren bir grafik çizilmiştir.

```python
plt.plot(setosa.Id, setosa.PetalLengthCm, color="blue", label="Setosa")
plt.plot(virginica.Id, virginica.PetalLengthCm, color="red", label="Virginica")
plt.plot(versicolor.Id, versicolor.PetalLengthCm, color="orange", label="Versicolor")
plt.legend() # Grafikte bulunan çizgilerin neyi temsil ettiğini gösterir.
plt.xlabel("Id") # X ekseninin etiketi
plt.ylabel("Petal Length (cm)") # Y ekseninin etiketi
plt.title("Setosa Sınıfının Görselleştirilmesi") # Grafiğin başlığı
plt.show()
```

### Seaborn kütüphanesinin **pairplot** fonksiyonu çok değişkenli verilerin dağılımını ve ilişkilerini görselleştirmek için kullanılan güçlü bir araçtır. Bu fonksiyon, veri setindeki tüm sayısal değişkenler arasındaki ilişkileri bir araya getirerek çiftler halinde grafikler oluşturur.

### Aşağıdaki kodda iris çiçek türlerinin bütün uzunluklarının birbirleriyle olan ilişkisi grafikle çizilmiştir.

```python
sns.pairplot(yeniVeri1, hue="Species")
plt.show()
```

### **Pairplot** fonksiyonundan elde edilen grafiklerden birini seçip grafiğini çizdirmek için **scatterplot** fonksiyonu kullanılabilir.

```python
sns.scatterplot(data=veri, x="SepalLengthCm", y="PetalLengthCm", hue="Species")
plt.show()
```

### Veriler analiz edilip görselleştirildikten sonra bu verilerin birbirleriyle olan ilişkilerine yani **Korelasyon** verilerine bakmamız gerekiyor.

### Bir veri setinin korelasyonunu bulmak için **corr** fonksiyonunu kullanabiliriz. Korelasyon 1'e yaklaştıkça verilerin aralarındaki pozitif ilişki artar.

```python
yeniVeri2 = yeniVeri1.drop(columns=['Species']) # Sayısal değerler üzerinde korelasyon olacağı için veri setinden sayısal olmayan değerler çıkartılır.
yeniVeri2.corr()
```

### Bu elde edilen korelasyonu görselleştirmek de faydalı olacaktır. Görselleştirme için **Seaborn** kütüphanesinin **heatmap** fonksiyonuyla ısı haritası çizdirebiliriz.

### Isı haritasında Koyu renkliler korelasyonun kötü olduğunu açık renkliler ise korelasyonun iyi olduğunu gösterir.

```python
korelasyon = yeniVeri2.corr()

sns.heatmap(korelasyon, annot=True) # annot, ısı haritasında her bir hücreye değerlerin yazılması için kullanılır.
plt.show()
```
